### Data Analysis and Visualisation

Statistical measure of the Dataset

There is quite a bit of variation between means and standard deviations implying we should consider standardising variables in pre-modelling.

```python
df.describe()
```

```python
#Number of values for each unique qualities

sns.countplot(x='quality', data=df)
```

```python
# Relation between volatile acidity and Quality

sns.barplot(x='quality', y='volatile acidity', data=df)
```

```python
sns.boxplot(x='quality', y='volatile acidity', data=df)
```

This graph shows that with the increase in quality, the volatile acidity decreases. This is due to decrease in acetic acid content, the wine gets more pleasant.

```python
# Relation between sulphates and Quality

sns.barplot(x='quality', y='sulphates', data=df)
```

```python
sns.boxplot(x='quality', y='sulphates', data=df)
```

This indicates that wine with lower acidity need more sulphates.

```python
# Relation between citric acid and Quality

sns.barplot(x='quality', y='citric acid', data=df)
```

```python
sns.boxplot(x='quality', y='citric acid', data=df)
```

This graph shows that the quality of wine increases as if we add preservative to the wine. It makes the wine more flavoured and freshness.

### Correlation

1. Positive Correlation : When both variables are directly proportional to each other.
2. Negative Correlation : When one variable is inversely proportional to second variable.

```python
correlation = df.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(), annot=True)
```

The following features are relatively correlated: total sulfur dioxide with free sulfur dioxide; fixed acidity with density and citric acid; alcohol with quality.

The following features are inversely correlated: fixed acidity with pH; citric acid with pH and volatile acidity.
